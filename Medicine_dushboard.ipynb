### Функция для обработки остальных таблиц

Все остальные таблицы имеют одинаковую структуру и могут быть объединены с добавлением столбца с локализацией опухоли. Для ускорения обработки напишем функцию.

```python
# пишем функцию для чтения и преобразования файлов
def convert_xlsx(table_name):
    xlsx_file = pd.ExcelFile(table_name) 
    sheet_names = xlsx_file.sheet_names #считываем название листов в файле
    sheet_1 = pd.read_excel(table_name, sheet_name=sheet_names[0]) #загружаем данные с первого листа
    sheet_2 = pd.read_excel(table_name, sheet_name=sheet_names[1]) #загружаем данные со второго листа
    # выбераем название локализации из шапки
    tab_name = sheet_1.columns[0].rstrip(' Таблица').split('\n')[1].split(' Т')[0].strip()
    # добавляем название на первый лист
    sheet_1.insert(0, "название таблицы", tab_name)
    # переименовываем столбцы на обеих листах
    sheet_1.columns = col_name
    sheet_2.columns = col_name_2
    # удаляем ненужные строки из шапки
    sheet_1.drop([0,1], inplace=True)
    sheet_2.drop([0,1,2], inplace=True)
    # удалим все строки с суммами по округам
    sheet_1 = sheet_1.query('область not in @districts and область != "РОССИЯ"')
    sheet_2 = sheet_2.query('область not in @districts and область != "РОССИЯ"')
    # добавим столбец с округом на первый лист
    sheet_1 = sheet_1.merge(regions)
    # перемещаем столбец с округом в начало
    temp_cols = sheet_1.columns.tolist()
    new_cols = temp_cols[-1:] + temp_cols[:-1]
    sheet_1 = sheet_1[new_cols]
    # соединяем 2 листа в одну таблицу
    table = sheet_1.merge(sheet_2)
    return table
```

### Таблицы 26-52 (tab5-29) с разной локализацией

Теперь в цикле пройдемся нашей функцией для предобработки и объединим предобработанные таблицы в одну.

```python
# обновляем список файлов, оставляя только необработанные ранее
xlsx_names = names[5:30]
columns_name = ['федеральный округ',
            'название таблицы', 
            'область', 
            'взято на учет впервые в жизни', 
            'в т.ч.выявлены активно, %', 
            'на учете на конец года, абсолютное число',
            'на учете на конец года, на 100тыс.населения',
            'на учете 5 лет и более, абсолютное число',
            'на учете 5 лет и более, %  от сост. на учете',
            'Индекс накопления контингентов',
            'Летальность, %',
            'зарегистрировано ЗНО без учета посмертно', 
            'диагноз подтвержден морфологически, %', 
            'со стадией 1, %', 
            'со стадией 2, %',
            'со стадией 3, %',
            'со стадией 4, %',
            'стадия не установлена, %',
            'летальность в первый год с установки, %']
# создаем пустой дф, куда будем добавлять данные из всех обрабатываемых таблиц
main_table = pd.DataFrame(columns=columns_name)
# проходимся в цикле по именам таблиц и после предобработки добавляем их в один файл
for name in xlsx_names:
    new_table = convert_xlsx(name)
    main_table = pd.concat([main_table, new_table], ignore_index = True)
```
